```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from matplotlib.pyplot import figure, show, close, xlabel, ylabel, title
from scipy.constants import k, h, c,e
from scipy.optimize import curve_fit
```

```python
hdulist = fits.open('Reynier files/galaxy_wl.fits')
hdulist.info()
print()

hdr = hdulist[0].header
dat = hdulist[0].data
print(f"data shape: {dat.shape}")


fig=figure()
frame=fig.add_subplot(1,1,1)
frame.plot(np.stack(dat, axis=1),color="navy", label="Galaxy spectrum")
frame.set_title("Galaxy")
frame.grid(alpha=0.2)
show()
close(fig)

hdulist.close()
```
